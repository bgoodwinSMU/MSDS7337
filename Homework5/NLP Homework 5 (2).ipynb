2.	Extract noun phrase (NP) chunks from your reviews using the following procedure:

a.	In Python, use BeautifulSoup to grab the main review text from each link.  

b.	Next run each review text through a tokenizer, and then try to NP-chunk it with a shallow parser. 

c.	You probably will have too many unknown words, owing to proper names of characters, actors, and so on that are not in your working dictionary.
```python
#Part A

reviewList = []

for x in range(0,len(dat)):
    url = dat['ReviewURL'].values[x]
    response = get(url)
    #print(response.text[:250])
    html_soup = BeautifulSoup(response.text, 'html.parser')
    movie_content = html_soup.find_all('div', class_ = 'content')
    first_movie = movie_content[0]
    review = [str(first_movie.div)]
    reviewList.append(review)
```

```python
#Part B
#Part C mixed in here as well
stop_words = set(stopwords.words('english'))
grammar = r"""
  NP: {<DT|PP\$>?<JJ>*<NN>}   
      {<NNP>+}                
      {<DT>?<JJ.*>*<NN.*>+}
      {<PRP$><NNP>+<NN>}
      {<DT><NN><VBG>}
      {<CD><NN><TO><CD><NN>}
      {<JJR><IN><CD><NN>}
      {<DT><JJS><VBG><NNS>}
"""
cp = nltk.RegexpParser(grammar)
tokenList = []
results = []


for x in range(0, len(dat)):
    reviewList[x] = str(reviewList[x])
    


for y in range(0,len(dat)):
    print("Review:")
    print("")
    text = nltk.word_tokenize(reviewList[y])
    filtered_review = [w for w in text if not w.lower() in stop_words]
    output = nltk.pos_tag(filtered_review)
    result = nltk.chunk.ne_chunk(output)
    cp = nltk.RegexpParser(grammar)
    result = cp.parse(result)
    results.append(result)
    for subtree in result.subtrees():
        if subtree.label() == 'NP': print(subtree)
    
    
    
```
